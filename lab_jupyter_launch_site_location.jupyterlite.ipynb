First, let's try to add each site's location on a map using site's latitude and longitude coordinates


The following dataset with the name `spacex_launch_geo.csv` is an augmented dataset with latitude and longitude added for each site.


```python
# Download and read the `spacex_launch_geo.csv`
from js import fetch
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df=pd.read_csv(spacex_csv_file)
```

Now, you can take a look at what are the coordinates for each site.


```python
# Select relevant sub-columns: `Launch Site`, `Lat(Latitude)`, `Long(Longitude)`, `class`
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]#This line filters the spacex_df DataFrame to only include the columns ‘Launch Site’, ‘Lat’ (latitude), ‘Long’ (longitude), and ‘class’.
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()#groups the data by the ‘Launch Site’ column and uses the .first() method to get the first occurrence of each group. The as_index=False parameter keeps ‘Launch Site’ as a column rather than turning it into the index. The resulting DataFrame, launch_sites_df, will have one row for each launch site with the latitude, longitude, and class of the first record in the original DataFrame for that site.
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]#further trims the launch_sites_df DataFrame to only include the ‘Launch Site’, ‘Lat’, and ‘Long’ columns, effectively removing the ‘class’ column from the previous step.
launch_sites_df
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN', 
    lng_first=False,
    num_digits=20,
    prefix='Lat:', #The text to display before the coordinates.
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0 

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1
#Compute the components of the haversine formula:
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c #Compute the distance by multiplying the angular distance by the Earth’s radius.
    return distance #The function returns the distance calculated between the two points. This method is known as the Haversine formula, which is particularly important in navigation.
```
